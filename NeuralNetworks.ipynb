```python
import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
[x_train, y_train], [x_test, y_test] = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train/255, x_test/255
```

## Multi-class Logistic Regression

```python
x = tf.keras.Input(x_train.shape[1:])
flat = tf.keras.layers.Flatten()(x)
y = tf.keras.layers.Dense(10, activation = 'softmax')(flat)

lr = tf.keras.Model(inputs = x, outputs = y)
lr.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
lr.summary()
```

```python
hist_lr = lr.fit(x_train, y_train, epochs = 100, batch_size = 256, validation_split = 0.2, verbose = False)
```

```python
plt.plot(hist_lr.history['accuracy'], label = 'train')
plt.plot(hist_lr.history['val_accuracy'], label = 'val')
plt.legend()
plt.show()
```

```python
hist_lr.history['accuracy'][-1], hist_lr.history['val_accuracy'][-1]
```

```python
lr.evaluate(x_test, y_test)
```
