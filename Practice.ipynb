```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Lecture3.xlsx")
df.head()
```

```python
def derivatives(df, w, b):
    df['yhat'] = w*df['Years Of Experience'] + b
    df['err'] = df['yhat'] - df['Annual Income']
    df['err_x'] = df['err']*df['Years Of Experience']
    df['SE'] = df['err']**2
    mse = df['SE'].sum()/(2*df['yhat'].count())
    derv_w = df['err_x'].sum()/df['yhat'].count()
    derv_b = df['err'].sum()/df['yhat'].count()

    return derv_w, derv_b, mse
```

```python
w, b = 0, 0
lr = 0.001
mse = [float('inf')]
eps = 1000
```

```python
while True:
    derv_w, derv_b, curr_mse = derivatives(df, w, b)
    w = w - lr*derv_w
    b = b - lr*derv_b
    if abs(curr_mse - mse[-1]) < eps: break
    mse.append(curr_mse)

print(w, b, len(mse) - 1)
```

```python
plt.plot(mse)
plt.show()
plt.plot(mse[1:])
plt.show()
plt.plot(mse[2:])
plt.show()
plt.plot(mse[3:])
plt.show()
```
